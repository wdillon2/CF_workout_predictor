```python
test = soup.find_all('div',{'class':'pull-left'})
movements = []
for s in test:
    x = s.find_all('div',{'class':'movement-stat stats-row clearfix'})
    for row in x:
        move = {}
        move['move_class'] = s.find('h3').text
        move['movement'] = row['data-name']
        move['frequency'] = int(row['data-count'])
        movements.append(move)

df = pd.DataFrame(movements)
df = df[['movement','frequency','move_class']]
```
